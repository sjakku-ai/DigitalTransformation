# **7. Model Building and Evaluation**
---

```python
le = LabelEncoder()
for i in df.columns:
    if df[i].dtype == "object":
        df[i] = le.fit_transform(df[i])
```

**Observation:**
Converts categorical variables to numeric format — required for machine learning.

```python
wcss=[]
K=range(1,13)
for i in K:
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

plt.plot(K,wcss,marker='o')
plt.title('Elbow Method',fontsize=15)
plt.xlabel('Number of clusters',fontsize=15)
plt.ylabel('Sum of Squared distance',fontsize=15)
plt.show()
```

**Observation:**
* Determines the optimal number of clusters (elbow point) using WCSS.
* Clusters dataset into 4 groups. Each row is assigned a cluster label.


```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
kmeans.fit(df)
```

**Observation:**
Clusters dataset into 4 groups. Each row is assigned a cluster label.

```python
pd.Series(kmeans.labels_).value_counts()
```

```python
labels = pd.DataFrame(kmeans.labels_)
labeled_data = pd.concat((df,labels),axis=1)
labeled_data = labeled_data.rename({0:'labels'},axis=1)
```

```python
labeled_data.head()
```

```python
fig = px.scatter(labeled_data, x= 'user_id',y = 'order_id', color = 'labels')
fig.show()
```

**Observation:**

* The x-axis represents different users (user_id), and the y-axis shows order IDs, representing individual purchases.

* The color='labels' argument uses the KMeans cluster assignments (0, 1, 2, 3) to group the data into visually distinguishable color-coded segments.

* This visualization helps identify patterns or similarities among customers based on their order behavior and characteristics derived from the dataset.






# **Implementation**

## **Cluster-wise Profile Analysis**

```python
# Step 1: Check numeric features to include in the analysis
numeric_features = df.select_dtypes(include=np.number).columns
print("Numeric Features:\n", numeric_features)

# Step 2: Group by cluster labels and calculate mean values
cluster_profile = labeled_data.groupby('labels')[numeric_features].mean()

# Step 3: Display as a transposed table for easier reading
cluster_profile = cluster_profile.T
print("Cluster-wise Feature Means:\n")
display(cluster_profile.style.background_gradient(cmap='YlGnBu'))
```

**Observation:**

Calculates the average value of each numeric feature (e.g., days since prior order, order hour, etc.) for each cluster.

Helps interpret user behavior — e.g., Cluster 0 may shop late at night, Cluster 1 orders frequently, etc.

## **Pick a real user from your dataset, find their feature vector, predict their cluster, and then recommend products based on what others in the same cluster purchased most frequently.**

```python
# Step 1: Pick an existing user
sample_user_id = labeled_data['user_id'].sample(1).values[0]  # Pick a random user
print(f"🎯 Sample User ID: {sample_user_id}")

# Step 2: Get that user's data (you can take mean if multiple entries)
user_data = labeled_data[labeled_data['user_id'] == sample_user_id]

# Drop the label column before prediction
features_for_prediction = user_data[df.columns].mean().values.reshape(1, -1)

# Step 3: Predict the cluster for this user
predicted_cluster = kmeans.predict(features_for_prediction)[0]
print(f"🔍 The user belongs to Cluster: {predicted_cluster}")
```

```python
# Get all users in the same predicted cluster
users_in_cluster = labeled_data[labeled_data['labels'] == predicted_cluster]

# Recommend Top 10 Most Frequently Bought Products in This Cluster
top_products = users_in_cluster['product_name'].value_counts().head(10)

print("\n📦 Recommended Products for This User:")
print(top_products)

# Optional: Visualize
plt.figure(figsize=(12, 6))
top_products.plot(kind='bar', color='green', edgecolor='black')
plt.title(f'Top Products in Cluster {predicted_cluster}')
plt.xlabel('Product Name')
plt.ylabel('Order Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Observation:**

This method makes your recommendation grounded in real, known behavior.

You can even apply this repeatedly to show personalized recommendations for multiple real users.

You now have a working clustering-based recommendation engine.
